# Introduction to Matplotlib

```python
It is a visualization library.
Can showcase data graphically by using the matplotlib.
```

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt 
```

```python
x = [1,2,3]
y = [1,4,9]

plt.plot(x,y)
plt.show()
```

```python
But now simple Python/Matplotlib, upto intermediate visualization stuff can be done freely.
```

```python
E.g.,

Revenue vs Sales

X-axis : Sales
Y-Axis : Revenue
    
Title of the graph --> Performance calculator of sales

```

```python
x = [1,2,3]
y = [1,4,9]

plt.title('Performance calculator of sales')
plt.xlabel('Sales')
plt.ylabel('Revenue')

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]

plt.title('Performance calculator of sales')
plt.xlabel('Sales')
plt.ylabel('Revenue')

plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]

plt.title('Performance calculator of sales')
plt.xlabel('Sales')
plt.ylabel('Revenue')

plt.plot(x,y)
plt.plot(x,z)
plt.legend(['this is xand y', 'this is x and z'])
plt.show()
```

```python
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]

plt.title('Performance calculator of sales')
plt.xlabel('Sales')
plt.ylabel('Revenue')


plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)
plt.legend(['this is xand y', 'this is x and z', 'this is y and z'])
plt.show()
```
